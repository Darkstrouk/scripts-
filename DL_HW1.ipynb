```python
model = nn.Sequential(

  nn.Linear(90, 50),
  nn.BatchNorm1d(50),
  nn.ReLU(),
  nn.Dropout(p=0.23),

  nn.Linear(50, 25),
  nn.BatchNorm1d(25),
  nn.ReLU(),
  nn.Dropout(p=0.35),

  nn.Linear(25, 1),
  nn.BatchNorm1d(1),
  nn.ReLU(),
  nn.Dropout(p=0.4)
)
```

```python
train_set = torch.utils.data.TensorDataset(X_train, y_train)
train_dataloader = torch.utils.data.DataLoader(train_set, batch_size=130, shuffle=True)

test_set = torch.utils.data.TensorDataset(X_test, y_test)
val_dataloader = torch.utils.data.DataLoader(test_set, batch_size=130, shuffle=True)


optimizer = torch.optim.Adam(model.parameters(), lr=0.01, betas=(0.9, 0.999)) 

criterion = lambda x, y: torch.sqrt(torch.mean((x-y) **2))

# loss_fn = nn.MSELoss()
# RMSE__train_loss = torch.sqrt(loss_fn(pridiction, target))
# criterion = RMSE_loss 
```
